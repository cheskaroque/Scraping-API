# Getting data from an API
This notebook walks you through some steps in collecting data from BBC News using the News API.


### Import package

```python
from newsapi import NewsApiClient
from datetime import datetime
from datetime import timezone

import pandas as pd
```

### Get articles via News API

```python
pd.set_option('display.max_colwidth', None)

newsapi = NewsApiClient(api_key = 'f563f261c4f741a28f330d7db0a312b5')

all_articles = newsapi.get_everything(sources = 'bbc-news',
                                     from_param = '2022-06-08',
                                     to = '2022-06-09',
                                     language = 'en',
                                     sort_by = 'popularity',
                                     )

newsArticles = all_articles.get('articles')
```

```python
news_list = pd.DataFrame(newsArticles)
```

```python
news_list.columns = ['source', 'author', 'title', 'description', 'url', 'urlToImage', 'Published','content' ]
```
